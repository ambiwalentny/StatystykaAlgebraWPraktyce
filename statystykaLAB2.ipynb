### Zadanie 1

```python
import statistics

# Definicja zmiennych losowych
X = [1, 2, 3, 4, 5, 6]
P = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]

# Obliczenie średniej
mean = statistics.mean(X)
print("Średnia:", mean)

# Obliczenie mediany
median = statistics.median(X)
print("Mediana:", median)

# Obliczenie mody
mode = statistics.mode(X)
print("Moda:", mode)

# Obliczenie odchylenia standardowego
std = statistics.stdev(X)
print("Odchylenie standardowe:", std)
```

### zadanie 2


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, poisson

# Ustawienia
n = 100

# Generowanie prób
bernoulli_samples = bernoulli.rvs(0.5, size=n)
binomial_samples = binom.rvs(20, 0.4, size=n)
poisson_samples = poisson.rvs(5, size=n)
```

### zadanie 3

```python
from scipy.stats import describe

# Statystyki podstawowe
bernoulli_stats = describe(bernoulli_samples)
binomial_stats = describe(binomial_samples)
poisson_stats = describe(poisson_samples)

print(bernoulli_stats)
print(binomial_stats)
print(poisson_stats)
```

### zadanie 4

```python
# Wykresy rozkładu prawdopodobieństwa
fig, axes = plt.subplots(3, 1, figsize=(8, 12))

# Rozkład Bernoulliego
unique, counts = np.unique(bernoulli_samples, return_counts=True)
axes[0].bar(unique, counts / n, align='center')
axes[0].set_xticks(unique)
axes[0].set_title('Rozkład Bernoulliego')

# Rozkład Dwumianowy
x_binomial = np.arange(0, 21)
axes[1].bar(x_binomial, binom.pmf(x_binomial, 20, 0.4), align='center')
axes[1].set_xticks(x_binomial)
axes[1].set_title('Rozkład Dwumianowy')

# Rozkład Poissona
x_poisson = np.arange(0, np.max(poisson_samples) + 1)
axes[2].bar(x_poisson, np.bincount(poisson_samples) / n, align='center')
axes[2].set_xticks(x_poisson)
axes[2].set_title('Rozkład Poissona')

plt.tight_layout()
plt.show()
```

### zadanie 6

```python
# Generowanie danych dla rozkładu normalnego
mean_normal = 0
std_normal = 2
data_normal = np.random.normal(mean_normal, std_normal, size=100)

# Statystyki podstawowe
stats_normal = describe(data_normal)

# Wyświetlanie wyników
print("Statystyki podstawowe dla rozkładu normalnego (średnia=0, odchylenie=2):")
print(f"Średnia: {stats_normal.mean}")
print(f"Wariancja: {stats_normal.variance}")
print(f"Kurtoza: {stats_normal.kurtosis}")
print(f"Skośność: {stats_normal.skewness}")

# Zwiększenie liczby danych
data_normal_large = np.random.normal(mean_normal, std_normal, size=1000)
stats_normal_large = describe(data_normal_large)

# Wyświetlanie wyników
print("\nStatystyki podstawowe dla większej próbki danych:")
print(f"Średnia: {stats_normal_large.mean}")
print(f"Wariancja: {stats_normal_large.variance}")
print(f"Kurtoza: {stats_normal_large.kurtosis}")
print(f"Skośność: {stats_normal_large.skewness}")
```
